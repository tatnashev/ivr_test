### Смотрим на тексты

Теперь посмотрим более детально на сами тексты. Для этого, в первую очередь, лемматизируем их 

```python
from nltk.corpus import stopwords 
nltk.download('stopwords')
```

Так как это обращения клиентов, попробуем хотя бы частично убрать приветствия и просто другие слова, не несущие особого смысла в данном случае

```python
additional_stopwords = ['здравствуйте', 'привет', 'это', 'очень']

stopwords = stopwords.words('russian') + additional_stopwords 
stopwords_set = set(stopwords)
```

```python
preprocessor = Preprocessor(stopwords_set)
```

Теперь можем предобработать наши тексты, а также сразу добавим еще один признак - количество лемм

*небольшой комментарий - возможно в данном случае использование класса для препроцессинга избыточно, однако, на мой взгляд, это достаточно удобная заготовка для того случая, если захочется сделать предобработку более сложной*

```python
data['preprocessed_text'] = data['text'].progress_apply(lambda x: preprocessor.preprocess_text(x)) 
data['len_preprocessed_text'] = data['preprocessed_text'].progress_apply(lambda x: len(x))
```

```python
y2label, target = preprocessor.preprocess_labels(data['label'])
data['target'] = target 
```

```python
plt.figure(figsize=(12, 7))
plt.title('Распределения длин текстов по классам')
sns.boxplot(data, x='len_preprocessed_text', y='label', order=class_distrib.index)
plt.show()
```

Явно бросается в глаза, что во всех классах в среднем у нас достаточно короткие тексты. Посмотрим на размеры словарей каждого класса

```python
def get_dict_by_label(data, label): 
    unique_lemmas = set()
    df_label = data[data['label'] == label] 
    
    for text in df_label['preprocessed_text']: 
        for lemma in text: 
            unique_lemmas.add(lemma)
            
    return unique_lemmas
```

```python
labels = class_distrib.index
label2unique_lemmas = defaultdict(set) 

for label in tqdm(labels): 
    label2unique_lemmas[label] = get_dict_by_label(data, label)
```

```python
label2count_lemmas = {label: len(unique_lemmas) for label, unique_lemmas in label2unique_lemmas.items()} 
df_label2count = pd.DataFrame.from_dict(label2count_lemmas, orient='index') 
df_label2count = df_label2count.reset_index().rename(columns={'index': 'label', 0 : 'count'})
df_label2count
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=df_label2count, y='label', x='count') 
plt.title('Размер словарей по классам') 
plt.show()
```

В целом все достаточно логично - чем больше класс, тем больше словарь (за исключением класса `Баланс`)

Теперь посмотрим на пересечение словарей. Гипотеза в данном случае заключается в следующем: если пересечение маленькое, то в дальнейшем качество при использование более сложных моделей (например, BERT) будет расти несильно относительно бейзлайна 

```python
# эффективнее было бы предпосчитать словари, и потом их попарно сравнивать, 
# но кажется, что на таких объемах это некритично 

label2label_intersection = defaultdict(lambda: defaultdict(set)) 

for i in range(len(labels)): 
    for j in range(i+1, len(labels)): 
        label_first = labels[i] 
        label_second = labels[j] 
        
        dict_first = get_dict_by_label(data, label_first)
        dict_second = get_dict_by_label(data, label_second) 
        
        dict_all = dict_first | dict_second 
        intersection = dict_first & dict_second 
        
        intersection_percent = (len(intersection) / len(dict_all)) * 100
        label2label_intersection[label_first][label_second] = intersection_percent 
        label2label_intersection[label_second][label_first] = intersection_percent
```

```python
df_label2label = pd.DataFrame.from_dict(label2label_intersection, orient='index')
df_label2label = df_label2label[df_label2label.index].fillna(100)

plt.figure(figsize=(12, 8))
sns.heatmap(data=df_label2label, vmin=0, vmax=100)
plt.title('Пересечение словарей между классами (%)')
plt.show()
```

К сожалению, данный показатель ничего не дал - все словари примерно в равной степени пересекаются между друг другом

Отдельно можно заметить, что словари наименьших классов имеют наименьшее пересечение с другими классами, что достаточно логично - это просто объясняется их размером

## Обучаем модели

Для начала зафиксируем обучающую, валидационную и тестовую выборки

```python
from sklearn.model_selection import train_test_split
```

```python
df_train, df_test = train_test_split(data, test_size=0.5, random_state=42, stratify=data['target']) 
df_val, df_test = train_test_split(df_test, test_size=0.4, random_state=42, stratify=df_test['target']) 
```

### Бейзлайн

```python
# !wget https://storage.yandexcloud.net/natasha-navec/packs/navec_news_v1_1B_250K_300d_100q.tar
# !mv navec_news_v1_1B_250K_300d_100q.tar data/navec_news_v1_1B_250K_300d_100q.tar
```

```python
vectorizer = Vectorizer(embeddings_path='data/navec_news_v1_1B_250K_300d_100q.tar') 
vectorizer.fit_tfidf(data['preprocessed_text'].apply(lambda x: ' '.join(x))) 
```

```python
df_train['embedding'] = df_train['preprocessed_text'].progress_apply(lambda x: vectorizer.get_embedding(x))
df_val['embedding'] = df_val['preprocessed_text'].progress_apply(lambda x: vectorizer.get_embedding(x))
df_test['embedding'] = df_test['preprocessed_text'].progress_apply(lambda x: vectorizer.get_embedding(x))
```

```python
X_train = df_train['embedding'] 
y_train = df_train['target'] 

X_test = df_test['embedding'] 
y_test = df_test['target']
```

```python
clf = LogisticRegression(random_state=42, max_iter=250)
clf.fit(np.vstack(X_train.values), y_train)
```

```python
y_pred = clf.predict(np.vstack(X_test.values))
```

```python
print(classification_report(y_test, y_pred))
```

### Бейзлайн + длина текста

```python
def concat_embeddings_and_number(data, embedding_col='embedding', number_col='len_preprocessed_text'): 
    embeddings = np.vstack(data[embedding_col]) 
    numbers = data[number_col].values.reshape(-1, 1) 
    
    return np.hstack((embeddings, numbers)) 
```

```python
X_train = concat_embeddings_and_number(df_train)
y_train = df_train['target'] 

X_test = concat_embeddings_and_number(df_test)
y_test = df_test['target']
```

```python
clf = LogisticRegression(random_state=42)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

Видим, что идея с использованием длины текста в качестве дополнительной фичи не сработала, поэтому переходим к другим моделям

### Заменяем логрег на catboost

```python
from catboost import Pool, CatBoostClassifier
```

```python
train_dataset = Pool(data=np.vstack(df_train['embedding']),
                     label=df_train['target']
                    )

test_dataset = Pool(data=np.vstack(df_test['embedding']),
                    label=df_test['target'] 
                   )

model = CatBoostClassifier(iterations=1000,
                           learning_rate=0.5, 
                           depth=2, 
                           loss_function='MultiClass', 
                           verbose=50
                          )

model.fit(train_dataset)
```

```python
y_pred = model.predict(test_dataset) 

print(classification_report(y_test, y_pred))
```

Получили качество чуть получше, чем в бейзлайне. Теперь перейдем к трансформерам (это в другом ноутбуке)
